# Clean trails dataframe for machine learning

```python
import pandas as pd
import csv
import numpy as np
import re
```

```python
# Import trails data

trails_df = pd.read_csv('Resources/trails_final.csv')
trails_df.set_index('trail_id', inplace=True)
trails_df.head()
```

```python
# Isolate the features column

features = trails_df['features'].astype(str)

# remove the brackets,'', and spaces to create a string of comma separated text

features = features.str.replace(r'(\[)', '')
features = features.str.replace(r'(\])', '')
features = features.str.replace(r'(\')', '')
features = features.str.replace(r'(\s)', '')
features
```

```python
# split column of lists into multiple columns of True/False values

features_list = []
for row in features:
    # make a list of all the possible available features

    for word in row.split(','):
        if word is '':
            continue
        if word not in features_list:
            features_list.append(word)
            
# make the features Series into a DataFrame

features_df = pd.DataFrame(features)

# then add each feature as a column in features_df w/ True or False if it is contained in the trail's features

for item in features_list:
    
    features_df[item] = features_df['features'].str.contains(item)

# Drop the original features column    
features_df = features_df.drop('features', axis=1)    
features_df
```

```python
# Isolate the activities column

activities = trails_df['activities'].astype(str)

# remove the brackets, spaces, and '' to create a string of comma separated text

activities = activities.str.replace(r'(\[)', '')
activities = activities.str.replace(r'(\])', '')
activities = activities.str.replace(r'(\')', '')
activities = activities.str.replace(r'(\s)', '')
activities
```

```python
# split column of lists into multiple columns of True/False values

activities_list = []
for row in activities:
    # make a list of all the possible available features

    for word in row.split(','):
        if word is '':
            continue
        if word not in activities_list:
            activities_list.append(word)
            
# make the features Series into a DataFrame

activities_df = pd.DataFrame(activities)

# then add each feature as a column in features_df w/ True or False if it is contained in the trail's features

for item in activities_list:
    
    activities_df[item] = activities_df['activities'].str.contains(item)
    
# Drop the original activities column

activities_df = activities_df.drop('activities', axis=1)    
activities_df
```

```python
# drop the old ['features', 'activities'] columns from the trails_df

bad_cols = ['_geoloc','visitor_usage','features','activities','units']

trails_df = trails_df.drop(columns=bad_cols, axis=1)

trails_df
```

```python
# merge the trails and features dataframes

detailed_trails_df = trails_df.merge(features_df, how='left', on='trail_id')
detailed_trails_df = detailed_trails_df.merge(activities_df, how='left', on='trail_id')
detailed_trails_df
```

```python
# Write detailed_trails_df to CSV:

merged_data_to_load = "./Resources/detailed_trails.csv"
detailed_trails_df.to_csv(merged_data_to_load, index=True)
```
